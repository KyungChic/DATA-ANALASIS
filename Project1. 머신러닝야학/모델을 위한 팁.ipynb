```
#기존에 Boston 프로젝트 값 불러오기###################################################################################################
```

```
import pandas as pd
import tensorflow as tf

path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(path)
```

```
indep = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax','ptratio', 'b', 'lstat']]
dep = boston[['medv']]

print(indep.shape, dep.shape)
```

```
x1 = tf.keras.layers.Input(shape = [13])

H1 = tf.keras.layers.Dense(8, activation = 'swish')(x1)
H1 = tf.keras.layers.Dense(8, activation = 'swish')(H1)
H1 = tf.keras.layers.Dense(8, activation = 'swish')(H1)

y1 = tf.keras.layers.Dense(1)(H1)

model1 = tf.keras.models.Model(x1,y1)
model1.compile(loss = 'mse')
```

```
model1.fit(indep, dep, epochs = 1000, verbose = 0)
model1.fit(indep, dep, epochs = 10)
```

```
model1.predict(indep[0:5])
```

```
####################################################################################################################################

# loss값이 일정구간에서 더이상 내려가지 않음을 알 수 있다.
```

```
# BatchNormalization 이용하기

## activation 함수를 분리 한 후 적용해주는 게 효과가 높다
```

```
x2 = tf.keras.layers.Input(shape = [13])

H2 = tf.keras.layers.Dense(8)(x2)
H2 = tf.keras.layers.BatchNormalization()(H2)
H2 = tf.keras.layers.Activation('swish')(H2)

H2 = tf.keras.layers.Dense(8)(H2)
H2 = tf.keras.layers.BatchNormalization()(H2)
H2 = tf.keras.layers.Activation('swish')(H2)

H2 = tf.keras.layers.Dense(8)(H2)
H2 = tf.keras.layers.BatchNormalization()(H2)
H2 = tf.keras.layers.Activation('swish')(H2)

y2 = tf.keras.layers.Dense(1)(H2)

model2 = tf.keras.models.Model(x2,y2)
model2.compile(loss = 'mse')
```

```
model2.fit(indep, dep, epochs=1000, verbose = 0)
model2.fit(indep, dep, epochs=10)
```

```
#iris 모델에도 적용 가능하다.
```

```
import tensorflow as tf
import pandas as pd

path = "https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv"
iris = pd.read_csv(path)

print(iris.head())
print(iris.columns)
```

```
#원핫인코딩
iris = pd.get_dummies(iris)
print(iris.columns)
```

```
indep = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
dep =  iris[['품종_setosa', '품종_versicolor', '품종_virginica']]

print(indep.shape)
print(dep.shape)
```

```
X = tf.keras.layers.Input(4)

H = tf.keras.layers.Dense(10, activation='swish')(X)
H = tf.keras.layers.Dense(10, activation='swish')(H)
H = tf.keras.layers.Dense(10, activation='swish')(H)

Y = tf.keras.layers.Dense(3, activation='softmax')(H)

model = tf.keras.models.Model(X, Y)
model.compile(loss = 'categorical_crossentropy', metrics = 'accuracy')

model.summary()
```

```
model.fit(indep, dep, epochs=1000, verbose=0)
model.fit(indep, dep, epochs=10)
```

```
print(model.predict(indep[-5:]))
print(dep[-5:])
```

```
#히든레이어 없이는 약 300번의 학습을 통해 정확한 예측(정확성 95%)
#노드 10개의 히든레이어 3개 추가한 학습 결과 : epochs=100의 fit 1회만으로 정확성 98% 확보
```

```
#하지만 여전히 100%까지 올라가지 않음을 알 수 있다.
```

```
## BatchNormalization 적용
```

```
BX = tf.keras.layers.Input(shape=[4])

BH = tf.keras.layers.Dense(10)(BX)
BH = tf.keras.layers.BatchNormalization()(BH)
BH = tf.keras.layers.Activation('swish')(BH)

BH = tf.keras.layers.Dense(10)(BH)
BH = tf.keras.layers.BatchNormalization()(BH)
BH = tf.keras.layers.Activation('swish')(BH)

BH = tf.keras.layers.Dense(10)(BH)
BH = tf.keras.layers.BatchNormalization()(BH)
BH = tf.keras.layers.Activation('swish')(BH)

BY = tf.keras.layers.Dense(3, activation='softmax')(BH)

Bmodel = tf.keras.models.Model(BX, BY)
Bmodel.compile(loss = 'categorical_crossentropy', metrics = 'accuracy')

Bmodel.summary()
```

```
Bmodel.fit(indep, dep, epochs=1000, verbose=0)
Bmodel.fit(indep, dep, epochs=10)
```

```
#loss값도 32bp로, accuracy값도 100%로 향상
```
