```
import pandas as pd
import tensorflow as tf
```

```
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(path)
print(iris.head())
print(iris.columns)
```

```
#원핫인코딩 : 범주형자료인 품종을 숫자형자료로 변환하는 과정
incoding = pd.get_dummies(iris)
print(incoding.head())
print(incoding.columns)
```

```
#독립/종속변수 값 불러오기
indep = incoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
dep = incoding[['품종_setosa', '품종_versicolor',
       '품종_virginica']]

print(indep.shape)
print(dep.shape)
```

```
x = tf.keras.layers.Input(shape=[4])
y = tf.keras.layers.Dense(3, activation='softmax')(x)
model = tf.keras.models.Model(x, y)
model.compile(loss = 'categorical_crossentropy',
              metrics = 'accuracy')
```

```
model.fit(indep, dep, epochs=1000, verbose=0)
```

```
#상단 5개 자료를 통한 검증
print(model.predict(indep[0:5]))
print(dep[0:5])
```

```
#하단 5개 자료를 통한 검증
print(model.predict(indep[-5:]))
print(dep[-5:])
```

```
#가중치 산출 (각 품종에 대한 가중치가 행렬의 형태로 나와있으며, 첫 열은 setosa / 두번째열은 versicolor / 세번째 열은 virginica 에 대한 확률값 산출에 대한 가중치가 나옴.)
model.get_weights()
```
