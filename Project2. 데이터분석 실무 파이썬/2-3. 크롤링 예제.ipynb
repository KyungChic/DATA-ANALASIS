```python
#BeautifulSoup import하기

from bs4 import BeautifulSoup
```

```python
#melon 사이트 열기

url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser') #HTML 구문을 분석하여 분류해줌.
```

```python
#100개의 노래 태그 찾기

##먼저, melon 사이트에서 개발자도구를 사용하여 한 노래가 가지는 태그를 찾기  --> 각 노래는 tr 태그 아래 있음을 알 수 있음.

songs = soup.select('table > tbody > tr')
print(len(songs))
print(songs[0])
```

```python
#<CF 검색하는 범위는 구체적일수록 좋음 : 'tr'로 크롤링할 경우 101개의 태그가 검색됨. (노래 위에 범주부분도 tr 태그)>


songs1 = soup.select('tbody > tr')

for song in songs1:
    print(song)
    print('\n')



songs2 = soup.select('tr')

for song in songs2:
    print(song)
    print('\n')
```

```python
# I.구체적으로 정보 찾아보기 (1. 한 곡에 대한 정보 찾기)

song = songs[0]  #1위곡 따오기
```

```python
##step 1. 'a' 태그 개수 찾아보기 --> 6개가 나옴. 

title = song.select('a')

len(title)
```

```python
##step 2. 'a'의 상위 범주인 'span' 추가 --> 2개가 나옴

title = song.select('span > a')
len(title)
```

```python
##step 3. 'span'의 상위 범주인 'div'(ellipsis.rank01) 추가 -- 1개로 좁혀짐
###띄어쓰기는 '.'으로 표기하는 것 같다

title = song.select('div.ellipsis.rank01 > span > a')
len(title)
```

```python
#하이퍼링크의 텍스트 따오기 --> Celebrity 로 맞음을 확인

print(title[0].text)
```

```python
#가수정보찾기

singer = song.select('div.ellipsis.rank02 > a')
print(len(singer)) #1개임을 확인

singer = song.select('div.ellipsis.rank02 > a')[0].text
print(singer)
```

```python
## II. 100 곡에 대한 정보 찾기

for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    print(title, singer, sep = '!')
    
```

```python
#version1. selenium + BeautifulSoup

from selenium import webdriver
driver = webdriver.Chrome(r'C:\Users\Gargantua\Desktop\files\chromedriver')    #driver 실행
driver.get("https://www.melon.com/chart/index.htm") #사이트 실행


from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')  #html 내려 받아 parser 하기


songs = soup.select('table > tbody > tr') #노래에 해당하는 부분 크롤링

for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singers = song.select('div.ellipsis.rank02 > a')
    
    singer = ''
    for temp in singers:                           ###가수 여러명일 때 응용
        if temp != singers[-1]:
            singer = singer + temp.text + ', '
        else:
            singer = singer + temp.text
            
    
    
    
    print(title, singer, sep = '!')
    
```

```python
#version2. selenium만 이용하기

from selenium import webdriver
driver = webdriver.Chrome(r'C:\Users\Gargantua\Desktop\files\chromedriver')
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)


songs = driver.find_elements_by_css_selector('table > tbody > tr')
for song in songs:
    title = song.find_elements_by_css_selector('div.ellipsis.rank01 > span > a')[0].text
    
    singers = song.find_elements_by_css_selector('div.ellipsis.rank02 > a')
    singer = ''
    
    for temp in singers:
        if temp != singers[-1]:
            singer = singer + temp.text + ', '
        else:
            singer = singer + temp.text
    
    print(title, singer, sep = ", ")
```
